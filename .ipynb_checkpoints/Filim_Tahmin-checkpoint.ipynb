```python
import pandas as pd
import numpy as np
```

```python
kolonlar=["user_ıd","item_id","rating","timestamp"]
df=pd.read_csv("users.data",sep="\t",names=kolonlar)
df.head()
```

```python
len(df)
```

```python
movie=pd.read_csv("movie_id_titles.csv")
movie.head()
```

```python
len(movie)
```

```python
df=pd.merge(df,movie,on="item_id")
df
```

```python
matrix=df.pivot_table(index="user_ıd",columns="title",values="rating")
matrix
```

```python
matrix.shape
```

```python
type(matrix)
```

```python
starwars=matrix[("Star Wars (1977)")]
starwars
```

```python
starwars.isnull().sum()
```

```python
starwars_corrwith = matrix.corrwith(starwars)
starwars_corrwith
```

```python
starwarscorr = pd.DataFrame(starwars_corrwith, columns=["correlasyon"])
starwarscorr.dropna(inplace=True)
starwarscorr.head()
```

```python
sıra = starwarscorr.sort_values("correlasyon", ascending=False).head(10)

sıra
```
